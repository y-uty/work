```python
# アンケート結果の読み込み
qres_org = pd.read_excel("nisshahp.xlsx")
```

```python
qres_org.head(5)
```

```python
# まずは、hp_opinionだけでやる
qres_h = qres_org[["no", "hp_understand", "hp_use", "hp_amount", "hp_opinion"]]
qres_h
```

```python
# 欠損値有無の確認
qres_h.isnull().sum()
```

```python
# 欠損値の補正：今回は、自由記入欄が空欄のアンケートは分析対象から外すこととする
qres_h = qres_h.dropna()
qres_h.isnull().sum()
```

```python
qres_h
```

```python
# 記号など、集計時に重要でない文字列を除去する(replace)
qres_h["hp_opinion"] = qres_h["hp_opinion"].str.replace("・", "")
qres_h["hp_opinion"] = qres_h["hp_opinion"].str.replace("、", "")
qres_h["hp_opinion"] = qres_h["hp_opinion"].str.replace("\\n", "")
```

```python
# どれくらいの文章量を書いてくれているか把握する
qres_h["hp_oplen"] = qres_h["hp_opinion"].str.len()
```
